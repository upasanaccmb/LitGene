```python
from shapleyAnalysis import *
import sys
import os
import pandas as pd
```

```python
clusters_path = 'solubilityNEW/enrichment2/'
dataset_name = 'all'
path_name = 'solubilityNEW/clean_genes_solubilityNEW.pkl'
shap_values = pickleLoad(path_name)
input_data_path = "clean_genes.csv"
gene_loaded_data = pd.read_csv(input_data_path)
sentences = gene_loaded_data["Summary"].tolist()
geneNames = gene_loaded_data["Gene name"].tolist()
gene_to_idx = {gene.strip():idx for idx, gene in enumerate(geneNames)}
```

```python
clusters_indexes = []
save_dir = 'solubilityNEW/shapValues_clusters2/'
# nClusters = len(os.listdir(clusters_path))
nClusters = 46
for c in range(nClusters):
    file_path = clusters_path+'GeneLLM_all_cluster'+str(c)+'.txt'
    with open(file_path,'r') as clusterfile:
        cluster = [line.strip() for line in clusterfile.readlines()]
    idxs = [gene_to_idx[g] for g in cluster if g in gene_to_idx]
    print(len(cluster)-len(idxs))
    shap_values_dict_temp = {'data':[],'values':[],'base_values':[]}
    for i in idxs:
        shap_values_dict_temp['data'].append(shap_values.data[i])
        shap_values_dict_temp['values'].append(shap_values.values[i])
        shap_values_dict_temp['base_values'].append(shap_values.base_values[i])
    explanation = shap.Explanation(values=shap_values_dict_temp['values'],data = shap_values_dict_temp['data'],base_values=shap_values_dict_temp['base_values'])
    pickleSave(explanation,save_dir,'cluster'+str(c)+'.pkl')
    clusters_indexes.append(idxs)
    pickleSave(idxs,save_dir,'cluster'+str(c)+'_indexes.pkl')
```
